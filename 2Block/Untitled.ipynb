3.Генерируется список случайных целых чисел. Определить, сколько в нем четных чисел, а сколько нечетных.

```python
import random
N = int(input())
even = 0
odd = 0
lst = [random.randint(-1000, 1000) for i in range(N)]
for i in range(N):
    if lst[i] % 2 == 0:
        even += 1
    else:
        odd += 1
print(f'Четных: {even}, Нечетных: {odd}')
```

4. Дан словарь, состоящий из пар слов. Каждое слово является синонимом к парному ему слову. Все слова в словаре различны. Заменить в строке все слова, входящие в словарь, как ключи, на их синонимы.

```python
synonyms = {
    "быстрый": "скорый",
    "храбрый": "смелый",
    "дом": "жилище",
    "красивый": "прекрасный",
    "машина": "автомобиль"
}

text = input().split()
for w in text:
    if w in synonyms:
        print(synonyms[w], end=" ")
    else:
        print(w, end=" ")

```

5. Напишите функцию fib(n), которая по данному целому неотрицательному n возвращает n-e число Фибоначчи. В этой задаче нельзя использовать циклы — используйте рекурсию.

```python
def fib(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)

print(fib(0))  # 0

print(fib(1))  # 1

print(fib(5))  # 5

print(fib(10)) # 55

    
```

6. Сосчитайте количество строк, слов и букв в произвольном текстовом файле. (слова разделены пробелом, \n не считается символом)

```python
def analyze(filename):
    with open(filename, "r") as f:
        text = f.read()

    num_lines = len(text.splitlines())
    num_words = len(text.split())
    num_letters = sum(1 for letter in text if not letter.isspace())

    return num_lines, num_words, num_letters


lines, words, letters = analyze("input.txt")
print("Строк:", lines)
print("Слов:", words)
print("Букв:", letters)
```

7. Создайте итератор, выводящий бесконечную геометрическую прогрессию. Параметры прогрессии задаются через аргументы генератора

```python
def geo_prog(start, ratio, count):
    term = start
    for _ in range(count):
        yield term
        term *= ratio

start = int(input("Введите первый элемент: "))
ratio = int(input("Введите знаменатель прогрессии: "))
count = int(input("Введите количество элементов: "))

print("Первые", count, "членов геометрической прогрессии:")
for value in geo_prog(start, ratio, count):
    print(value)


```
